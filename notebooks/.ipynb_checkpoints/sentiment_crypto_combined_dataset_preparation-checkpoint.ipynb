### import bitcoin dataset

```python
btcDf = pd.read_csv('bitcoin.csv')
btcDf = btcDf.filter(['Open time', 'Close'])
btcDf['Open time'] = pd.to_datetime(btcDf['Open time'], unit='ms')
btcDf.set_index('Open time')
btcDf.to_csv('bitcoin_filtered.csv',index=False)
btcDf
```

### we found some missing values in the bitcoin dataset. update manulally

```python
# missing values found and updated manually
missing = [['2020-02-19 12:00:00', '10073.97'],
 ['2020-02-19 13:00:00', '10178.98'],
 ['2020-02-19 14:00:00', '10136.26'],
 ['2020-02-19 15:00:00', '10142.80'],
 ['2020-02-19 16:00:00', '10161.32'],
 ['2020-03-04 10:00:00', '8842.29'],
 ['2020-04-25 02:00:00', '7504.71'],
 ['2020-04-25 03:00:00', '7493.00'],
 ['2020-06-28 02:00:00', '9048.48'],
 ['2020-06-28 03:00:00', '8963.34'],
 ['2020-06-28 04:00:00', '8972.63'],
 ['2020-11-30 06:00:00', '17317.65'],
 ['2020-12-21 15:00:00', '22697.42'],
 ['2020-12-21 16:00:00', '23111.98'],
 ['2020-12-21 17:00:00', '23173.91'],
 ['2020-12-25 02:00:00', '23409.70'],
 ['2021-02-11 04:00:00', '44796.97'],
 ['2021-03-06 02:00:00', '49056.77'],
 ['2021-04-20 02:00:00', '56306.44'],
 ['2021-04-20 03:00:00', '56047.96'],
 ['2021-04-25 05:00:00', '50909.15'],
 ['2021-04-25 06:00:00', '50369.18'],
 ['2021-04-25 07:00:00', '50088.30'],
 ['2021-08-13 02:00:00', '44170.01'],
 ['2021-08-13 03:00:00', '43882.45'],
 ['2021-08-13 04:00:00', '44419.43'],
 ['2021-08-13 05:00:00', '44507.59'],
 ['2021-09-29 07:00:00', '41543.64'],
 ['2021-09-29 08:00:00', '42199.99']]

# update the bitcoin filterd dataset file with proper data
import csv
data=[]
with open('bitcoin_filtered.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        data.append(row)

data.remove(['Open time', 'Close'])
for row in missing :
    data.append(row)
    
df = pd.DataFrame(data,columns=['Open time', 'Close' ])
df['Open time'] = pd.to_datetime(df['Open time'])
df.set_index('Open time')

df = df.sort_values(by='Open time')

df.to_csv('bitcoin_filtered.csv', index=False)
```

### load the updated bitcoin dataset

```python
btcDf = pd.read_csv('bitcoin_filtered.csv')
btcDf['Open time'] = pd.to_datetime(btcDf['Open time'])
btcDf.set_index('Open time')
print(btcDf)
btcDf.plot(x='Open time', y='Close')
```

### concatenate both of the datasets

```python
result = pd.concat([btcDf, sentiDf], axis=1, join='inner')
result = result.filter(['datetime', 'Close', 'sentiment'])
result.set_index('datetime')
df.to_csv('final_dataset.csv', index=False)
print(result)
```
