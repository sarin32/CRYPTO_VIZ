sentiment data to be taken from
https://senticrypt.com/

index.json of api conains all the json filename of sentiment data

```python
import requests
sentiment_index = requests.get('http://api.senticrypt.com/v1/history/index.json').json()
sentiment_index
```

### sample data

```python
data = requests.get('https://api.senticrypt.com/v1/history/bitcoin-2020-02-13_20.json')
data
```

the sentiment index contains some unwanted file names
find the index of last wanted file

```python
sentiment_index.index('bitcoin-2022-05-29_05.json')
```

```python
sentiment_data = []
for t_stamp in sentiment_index[0:20049+1]:
    date_str = t_stamp[8:21]
    mean = requests.get('http://api.senticrypt.com/v1/history/' + t_stamp).json()[0]['mean']

    sentiment_data.append([date_str, mean])
```

```python
sentiment_data
```

```python
import csv
with open('data1.csv', 'w') as f:
      
    # using csv.writer method from CSV package
    write = csv.writer(f)
    write.writerows(sentiment_data)
```
