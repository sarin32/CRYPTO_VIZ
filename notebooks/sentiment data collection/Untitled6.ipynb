```python
import requests
sentiment_index = requests.get('http://api.senticrypt.com/v1/history/index.json').json()
print('index loaded')
```

```python
import csv
sentiment_data = []
with open('data6.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        if row==[]:
            continue
        sentiment_data.append(row)
print(sentiment_data)
```

```python
s =sentiment_data[-1][0]
start = sentiment_index.index('bitcoin-'+s+'.json')+1
stop = sentiment_index.index('bitcoin-2021-08-13_20.json')

for t_stamp in sentiment_index[start:stop]:
    date_str = t_stamp[8:21]
    print(date_str)
    mean = requests.get('http://api.senticrypt.com/v1/history/' + t_stamp).json()[0]['mean']

    sentiment_data.append([date_str, mean])
```

```python
import csv
with open('data6.csv', 'w') as f:
      
    # using csv.writer method from CSV package
    write = csv.writer(f)
    write.writerows(sentiment_data)
```
